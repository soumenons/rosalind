```python
def complement_dna(dna):
    comp=""
    for base in dna:
        if base == 'A':
            comp+='T'
        elif base == 'C':
            comp+='G'
        elif base == 'G':
            comp+='C'
        elif base == 'T':
            comp+='A'
        else:
            print("Unknown base in string")
    comp=comp[::-1]
    print(comp)
```
